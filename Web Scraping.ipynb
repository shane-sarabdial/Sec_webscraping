```python
import requests
from bs4 import BeautifulSoup
```

```python
# define base url
base_url = r"https://www.sec.gov/Archives/edgar/data"

#define a CIK number to do company search
cik_num = '/886982/'
```

```python
# Create a filing url
filing_url = base_url + cik_num +'index.json'

#request url
headers = {'user-agent': 'sample text'}
content = requests.get(filing_url,headers=headers)
decoded_content = content.json()
```

```python
# go into directroy key then item then get first element then go to name key to get filing number
filing_num = decoded_content['directory']['item'][0]['name']

#define filing number url
filing_url = base_url + cik_num + filing_num +"/index.json"
```

```python
# request new url
content = requests.get(filing_url, headers=headers)
document_content = content.json()

#get document names
for document in document_content['directory']['item']:
    if document['type'] != 'image2.gif':
        doc_name = (document['name'])
        document_url = filing_url +'/' + doc_name
        print(document_url)
```
